```python
from scipy import stats

stats.pearsonr(iris.SepalLengthCm, iris.SepalWidthCm)
```

```python
plt.figure(figsize=(8,8))
seaborn.heatmap(iris.corr(), annot=True, cmap="coolwarm").set_title('Find_Corelation')
```

```python
#Matplotlib Correlation Matrix
```

```python
import matplotlib.pyplot as plt
plt.imshow(iris.corr(), cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.xticks(range(len(iris.columns)), iris.columns, rotation=90)
plt.yticks(range(len(iris.columns)), iris.columns)
plt.gcf().set_size_inches(8,8)

labels = iris.corr().values
for y in range(labels.shape[0]):
    for x in range(labels.shape[1]):
        plt.text(x, y, '{:.2f}'.format(labels[y, x]), ha='center', va='center', color='white')
```
